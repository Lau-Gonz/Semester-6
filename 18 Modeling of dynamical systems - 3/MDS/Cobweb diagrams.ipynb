```python
def linear_function(x, slope, intercept):
    return slope * x + intercept

def cobweb_plot(iterations, initial_x, slope, intercept):
    x_values = [initial_x]
    y_values = [linear_function(initial_x, slope, intercept)]

    for i in range(iterations):
        x_values.append(y_values[-1])
        y_values.append(linear_function(x_values[-1], slope, intercept))

        plt.plot([x_values[-2], x_values[-1]], [y_values[-2], y_values[-1]], color='b')  # Line segment
        plt.plot(x_values[-1], y_values[-1], 'ro')  # Endpoint of the line segment

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Cobweb Plot for Linear Function')
    plt.grid(True)
    plt.show()

# Parameters for the linear function: y = slope * x + intercept
slope = -0.6
intercept = 1.2

# Cobweb plot parameters
iterations = 4
initial_x = 1.5

cobweb_plot(iterations, initial_x, slope, intercept)
```
