# <FONT SIZE=4 COLOR="bronw"> 1. Función para quitar tildes </FONT>

```python
def quitar_tildes(text):
  text = re.sub('á', 'a', text)
  text = re.sub('é', 'e', text)
  text = re.sub('í', 'i', text)
  text = re.sub('ó', 'o', text)
  text = re.sub('ú', 'u', text)
  return text
```

```python
quitar_tildes("Fabián Sánchez accionó la cámara de este.")
```

# <FONT SIZE=4 COLOR="bronw"> 2. Dejar solo letras</FONT>

```python
def solo_texto(text):
  text = re.sub("[^A-Za-zü]+", " ", text)
  return text
```

```python
solo_texto("fabian? est¡ . estudiando# mucho4454")
```

# <FONT SIZE=4 COLOR="bronw"> 3. Dejar Alfanúmerico</FONT>

```python
def alfanumerico(text):
  text = re.sub("[^A-Za-z0-9ü]+", " ", text)
  return text
```

```python
alfanumerico("fabian? est¡ . estudiando# mucho4454")
```

# <FONT SIZE=4 COLOR="bronw"> 4. Convertir a minúscula</FONT>

```python
def minuscula(text):
  text = text.lower()
  return text
```

```python
minuscula("Fabian Sánchez estuvo es su Casa. Casa de lobos")
```

```python
stop_words=get_stop_words("es")
```

```python
def eliminar_SW(text):
  text_clean = [palabra for palabra in text.split() if (palabra not in stop_words)]
  text_clean = " ".join(text_clean)
  return text_clean
```

```python
eliminar_SW("Fabian esta en la solución de algunas cosas de etica y algo más")
```

```python
def limpieza(text):
  text = quitar_tildes(text)
  text = minuscula(text)
  text = solo_texto(text)
  text = eliminar_SW(text)
  return text
```

```python
limpieza("Fabián en la Sánchez atrás@ Salazar56# estuvo¡ trabajando.")
```
