```python
import nltk
nltk.download("all")
```

### <FONT SIZE=4 COLOR="ff5c8a"> a. Importar y leer la base de tweets

```python
# a. Importar y leer la base de tweets
import pandas as pd
tweets = pd.read_csv("/content/tweets_politica.csv", sep = '::::')
tweets.head()
```

### <FONT SIZE=4 COLOR="ff5c8a"> b. Agregar al data.frame que se obtiene una columna que indique el número de caracteres de cada tweet.



```python
# b. Agregar al data.frame que se obtiene una columna que indique el número de caracteres de cada tweet.

list_carac = []
for i in range(len(tweets)):
  if tweets.tweet[i] == None:
    list_carac.append(0)
  else:
    list_carac.append(len(tweets.tweet[i]))

tweets.insert(loc=4, column='Caracteres', value=list_carac)

tweets.head()
```

### <FONT SIZE=4 COLOR="ff5c8a"> c. Tomar la variable ":tweets" y dar respuesta a los siguientes items.

#### <FONT SIZE=3 COLOR="00b4d8"> Encontrar el top 20 de usuarios </FONT>

```python
# Encontrar los usuarios en un tweet

#import re
patron_1 = r'(@\w{2,})'
list_user = []

for i in range(len(tweets)):
  if tweets.tweet[i] == None:
    continue
  elif re.findall(patron_1, tweets.tweet[i]) == []:
    continue
  else:
    list_user.append(re.findall(patron_1, tweets.tweet[i]))
```

```python
# Aplanamos la lista de usuarios
list_user_f = []
for i in list_user:
  list_user_f += i

# Obtenemos las frecuencias de la repetición de usuarios
freq_user =nltk.FreqDist(list_user_f)
```

#### <FONT SIZE=3 COLOR="00b4d8"> Tabla de frecuencias usuarios</FONT>

```python
# Obtenemos la tabla de frecuencias del top 20 de usuarios
df_user = pd.DataFrame([[key, freq_user[key]] for key in freq_user.keys()], columns=['Palabra', 'Frecuencia'])
df_top_user = df_user.sort_values("Frecuencia", ascending=False)
df_top_user.head(20)
```

#### <FONT SIZE=3 COLOR="00b4d8"> Gráfico de barras </FONT>

```python
import plotly.express as px
px.bar(x= df_top_user.Palabra[0:20],
       y= df_top_user.Frecuencia[0:20])
```

#### <FONT SIZE=3 COLOR="00b4d8"> Gráfico de línea </FONT>

```python
px.line(x= df_top_user.Palabra[0:20],
       y= df_top_user.Frecuencia[0:20])
```

#### <FONT SIZE=3 COLOR="00b4d8"> Encontrar top 10 de hashtags </FONT>

```python
# Encontrar los hashtags en un tweet

#import re
patron_2 = r'(#\w{2,})'
list_topic = []

for i in range(len(tweets)):
  if tweets.tweet[i] == None:
    continue
  elif re.findall(patron_2, tweets.tweet[i]) == []:
    continue
  else:
    list_topic.append(re.findall(patron_2, tweets.tweet[i]))
```

```python
# Aplanamos la lista de topics
list_topic_f = []
for i in list_topic:
  list_topic_f += i

# Obtenemos las frecuencias de la repetición de topics
freq_topic =nltk.FreqDist(list_topic_f)
```

#### <FONT SIZE=3 COLOR="00b4d8"> Tabla de frecuencias hashtags </FONT>

```python
# Obtenemos la tabla de frecuencias del top 10 de topics
df_user = pd.DataFrame([[key, freq_topic[key]] for key in freq_topic.keys()], columns=['Palabra', 'Frecuencia'])
df_top_user = df_user.sort_values("Frecuencia", ascending=False)
df_top_user.head(10)
```

## <FONT SIZE=3 COLOR="purple"> PARTE 4 </FONT>

Defina una función de preprocesado que se llamará **cleanPLN** teniendo en cuenta:

- Pasar a minúsculas

- Remover caracteres especiales

- Remover tildes

- Tokenizar

- Eliminar las stop.words (español o inglés, según sea el caso)

- Dejar palabras con más de 3 caracteres

- Regrese la cadena de texto simplificada.

Haga un ejemplo para verificar su funcionamiento.

```python
!pip install stop-words
```

```python
# Función para preprocesado
import re
from stop_words import get_stop_words

def cleanPLN(texto, idioma = "es"):
  # Pasar a minusculas
  texto = texto.lower()
  # Remover caracteres especiales y tildes
  texto = re.sub(r'[á]+', 'a', texto)
  texto = re.sub(r'[é]+', 'e', texto)
  texto = re.sub(r'[í]+', 'i', texto)
  texto = re.sub(r'[ó]+', 'o', texto)
  texto = re.sub(r'[úü]+', 'u', texto)
  texto = re.sub(r'[^A-Za-z\s]+', '', texto)
  # Tokenizar
  tokens = nltk.word_tokenize(texto)
  # Eliminar las stop.words y Dejar palabras con más de 3 caracteres
  stop_words=get_stop_words(idioma)
  tokens_clean = [palabras for palabras in tokens if len(palabras)>3 and (palabras not in stop_words)]
  # Regrese la cadena de texto simplificada.
  texto_limpio = ' '.join(tokens_clean)
  return texto_limpio
```

**Ejemplo de prueba para la función cleanPLN**

```python
# Probemos la función
texto2 = "Gabriel García Márquez era un hombre de grandes principios. No solo morales. Los principios de sus novelas son de una belleza literaria, una capacidad de atracción y una perfección inigualables. Por eso millones de lectores se los han aprendido de memoria. El más conocido, claro está, es el archicitado comienzo de Cien años de soledad (1967): «Muchos años después, frente al pelotón de fusilamiento, el coronel Aureliano Buendía había de recordar aquella tarde remota en que su padre lo llevó a conocer el hielo. Macondo era entonces una aldea de 20 casas de barro y cañabrava construidas a la orilla de un río de aguas diáfanas que se precipitaban por un lecho de piedras pulidas, blancas y enormes como huevos prehistóricos. El mundo era tan reciente, que muchas cosas carecían de nombre, y para mencionarlas había que señalarlas con el dedo». A la misma altura, sin embargo, raya el inicio de Crónica de una muerte anunciada (1981), que tiene la dificultad superior de atraer con entusiasmo con un texto en el que en su primera línea conocemos ya el desenlace de la novela: «El día en que lo iban a matar, Santiago Nasar se levantó a las 5.30 de la mañana para esperar el buque en que llegaba el obispo». Y el tercero en la lista sería el comienzo de El amor en los tiempos del cólera (1985): «Era inevitable: el olor de las almendras amargas le recordaba siempre el destino de los amores contrariados»."
cleanPLN(texto2)
```
